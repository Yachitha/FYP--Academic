```python
api_get_users = 'https://api.github.com/users'


def call_api(apicall, **kwargs):

    data = kwargs.get('page', [])
    
    #make api call
    resp = requests.get(apicall, auth=(myEmail,myAccessToken))
    #add data into a json
    data += resp.json()

    # failsafe
    if len(data) > 3500:
        return (data)

    if 'next' in resp.links.keys():
        return (call_api(resp.links['next']['url'], page=data))

    return (data)
    
def csv_cvt(data):
    normalized=json_normalize(data)
    normalized.to_csv('ex2.csv', encoding='utf-8', index=False)
#     normalized.to_excel('ex3.xls', encoding='utf-8', index=False)

print("start api call")
data = call_api(api_get_users)
print("data collected!")
print("start adding to csv")
csv_cvt(data)
print("end of adding data to csv")
```
